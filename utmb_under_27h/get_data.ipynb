# Get data

```python
# UTMB API Tokens
utmb_live_access_token = get_utmb_live_access_token(
    utmb_username=utmb_website_login_username, 
    utmb_password=utmb_website_login_password
)

utmb_access_token = get_utmb_access_token(
    utmb_username=utmb_website_login_username, 
    utmb_password=utmb_website_login_password
)
```

```python
# Get & Save UTMB data for {2022, 2023, 2024, 2025}

save_path = os.path.join(workspace_root, "analysis", "utmb_under_", "utmb_race_results")
os.makedirs(save_path, exist_ok=True)
save_path_csv = os.path.join(save_path, "utmb_{}_race_results.parquet")

for year in [2022, 2023, 2024, 2025]:

    try:

        utmb_year_race_results_df = get_race_results(
            race_id = "utmb", 
            race_year = year, 
            course_id = "utmb", 
            access_token = utmb_live_access_token, 
            printouts = False

        )

        utmb_year_race_results_df.to_parquet(
            save_path_csv.format(year)

        )

        print(f"Data for year {year} fetched and saved successfully.")

    except Exception as e:

        print(f"Error fetching data for year {year}: {e}. Skipping...")
```
