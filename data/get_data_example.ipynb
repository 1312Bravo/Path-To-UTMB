### UTMB API Authentication Token - Sign in with UTMB credentials
- Live Access Token - Get Race Data

```python
# Get UTMB Sign in credentials (from environment variables)
from dotenv import load_dotenv
import os

load_dotenv()
utmb_website_login_username = os.getenv("utmb_website_login_username")
utmb_website_login_password = os.getenv("utmb_website_login_password")
```

```python
# UTMB Live API Token
utmb_live_access_token = get_utmb_live_access_token(
    utmb_username=utmb_website_login_username, 
    utmb_password=utmb_website_login_password
)
print(f"UTMB Live API Access Token: {utmb_live_access_token[:10]}...")
```

```python
# UTMB API Token
utmb_access_token = get_utmb_access_token(
    utmb_username=utmb_website_login_username, 
    utmb_password=utmb_website_login_password
)
print(f"UTMB API Access Token: {utmb_access_token[:10]}...")
```
